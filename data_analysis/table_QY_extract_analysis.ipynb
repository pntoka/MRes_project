```python
import os
import pandas as pd
import json
from pandas import json_normalize
import ast
```

```python
#importing file
# path = '/Users/pnt17/Library/CloudStorage/OneDrive-ImperialCollegeLondon/MRes_project_data/ceder_extract_data'
path = r'C:\Users\Piotr\OneDrive - Imperial College London\MRes_project_data\ceder_extract_data'
filename = 'table_QY_extract.json'
with open(os.path.join(path, filename), 'r', encoding = 'utf-8') as file:
    data = json.load(file)
```

```python
#creating dataframe
df = json_normalize(data)
df
```

```python
#dropping rows with no QY values
QY_df = df[df['QY'].apply(lambda x: len(x) > 0)]
QY_df
```

```python
qy_df_clean = pd.read_csv(os.path.join(path, 'qy_table.csv'))
qy_df_clean
```

```python
rxn_temp_time_df = pd.read_csv(os.path.join(path, 'clean_rxn_temp_time.csv'))
precursor_df = pd.read_csv(os.path.join(path, 'precursors_clean.csv'))
rxn_temp_time_df['rxn_temp_time'] = rxn_temp_time_df['rxn_temp_time'].apply(ast.literal_eval)
precursor_df['precursors'] = precursor_df['precursors'].apply(ast.literal_eval)
qy_df_clean['QY'] =qy_df_clean['QY'].apply(ast.literal_eval)
```

```python
common_doi = list(set(qy_df_clean['DOI'].to_list()) & set(rxn_temp_time_df['DOI'].to_list())&set(precursor_df['DOI'].to_list()))
print(len(common_doi))
common_doi
```

```python
qy_rxn_precursor_df = pd.DataFrame(columns = ['DOI', 'QY', 'rxn_temp_time', 'precursors'])
qy_rxn_precursor_df['DOI'] =common_doi
qy_dict = {}
rxn_temp_time_dict = {}
precursor_dict = {}
for doi in common_doi:
    qy = qy_df_clean.loc[qy_df_clean['DOI'] == doi, 'QY'].values
    rxn_temp_time = rxn_temp_time_df.loc[rxn_temp_time_df['DOI']==doi, 'rxn_temp_time'].values
    precursors = precursor_df.loc[precursor_df['DOI']==doi, 'precursors'].values
    qy_dict.update({doi:qy})
    rxn_temp_time_dict.update({doi:rxn_temp_time})
    precursor_dict.update({doi:precursors})
qy_rxn_precursor_df['QY'] = qy_rxn_precursor_df['DOI'].map(qy_dict)
qy_rxn_precursor_df['rxn_temp_time'] = qy_rxn_precursor_df['DOI'].map(rxn_temp_time_dict)
qy_rxn_precursor_df['precursors'] = qy_rxn_precursor_df['DOI'].map(precursor_dict)
```

```python
def flatten_list(values):
    flat_list = [item for sublist in values for item in sublist]
    return flat_list

qy_rxn_precursor_df['rxn_temp_time'] = qy_rxn_precursor_df['rxn_temp_time'].apply(flatten_list)
qy_rxn_precursor_df['QY'] = qy_rxn_precursor_df['QY'].apply(flatten_list)
qy_rxn_precursor_df['precursors'] = qy_rxn_precursor_df['precursors'].apply(flatten_list)
```
